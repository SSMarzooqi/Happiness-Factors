## Methodology

```python
# Importing necessary libraries for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns  
from scipy.stats import linregress

# Load the dataset from a CSV file
data = pd.read_csv("2019.csv")

# Drop the column 'Country or region' from the dataset and store the modified dataset in 'data_numeric'
data_numeric = data.drop(columns=['Country or region'])

# Perform data cleaning by removing rows with missing values
data_numeric.dropna(inplace=True)

# Convert non-numeric columns to numeric type if possible, coercing errors
data_numeric = data_numeric.apply(pd.to_numeric, errors='coerce')

data_numeric.head()
```

**Description of the Dataset:**
The dataset comprises socio-economic indicators from different countries along with subjective well-being scores obtained from surveys. It includes variables such as GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption. The following columns: GDP per Capita, Family, Life Expectancy, Freedom, Generosity, Trust Government Corruption describe the extent to which these factors contribute in evaluating the happiness in each country. If you add all these factors up, you get the happiness score so it might be un-reliable to model them to predict Happiness Scores.

**Corrolation Heatmap:** This part uses Matplotlib to create a heatmap and Seaborn to visualize the correlation between  columns in the dataset. It enhances the aesthetics of the heatmap with color mapping ('coolwarm'), annotating the cells with correlation values, and setting the title. Finally, it saves the plot as an image and displays it.

```python
# Create a heatmap to visualize the correlation between numeric columns in the dataset
sns.heatmap(data_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')  # Set title for the plot
plt.savefig('heatmap.png')  # Save the plot as an image
plt.show()  # Display the plot
```

**Scatter Plots** The next part is plotting the plots of the variables with most corrolation with the happiness score according to the previous heatmap. In addition to that the plot of healthy life expectancy and GDP per capita as the heatmap shows a sigificant corrolation between the two.

```python
# Create a scatter plot with regression line for 'GDP per capita' and 'Happiness Score'
sns.regplot(x='GDP per capita', y='Score', data=data_numeric)
plt.title('GDP per capita vs Happiness Score')  # Set title for the plot
plt.xlabel('GDP per capita')  # Set label for x-axis
plt.ylabel('Happiness Score')  # Set label for y-axis

# Calculate R-squared value
slope, intercept, r_value, p_value, std_err = linregress(data_numeric['GDP per capita'], data_numeric['Score'])
r_squared = r_value ** 2
plt.savefig('scatterplot GDP per capita.png')  # Save the plot as an image
plt.show()  # Display the plot
print('R-squared =' , round(r_squared, 2)) # Print the value of R-Squared
```

The regression line represents the linear relationship between the two variables. ## Results

```python
# Create a heatmap to visualize the correlation between numeric columns in the dataset
sns.heatmap(data_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')  # Set title for the plot
plt.savefig('heatmap.png')  # Save the plot as an image
plt.show()  # Display the plot
```

**Corrolation Heatmap** This corrolation heatmap shows that the variables that affect happiness scores the most are (GDP per capita, Social support, healthy life expectancy). It also shows that there is a signifacant corrolation between healthy life expectancy and GDP per capita.

```python
# Create a figure and axes
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot with regression line for 'GDP per capita' and 'Happiness Score'
sns.regplot(x='GDP per capita', y='Score', data=data_numeric, ax=axes[0])
axes[0].set_title('GDP per capita vs Happiness Score')  # Set title for the plot
axes[0].set_xlabel('GDP per capita')  # Set label for x-axis
axes[0].set_ylabel('Happiness Score')  # Set label for y-axis
# Calculate R-squared value
slope, intercept, r_value, p_value, std_err = linregress(data_numeric['GDP per capita'], data_numeric['Score'])
r_squared = r_value ** 2
axes[0].text(0.1, 0.9, f'R-squared: {r_squared:.2f}', transform=axes[0].transAxes, fontsize=12, bbox=dict(facecolor='lightgray', alpha=0.5))

# Scatter plot with regression line for 'Social Support' and 'Happiness Score'
sns.regplot(x='Social support', y='Score', data=data_numeric, ax=axes[1])
axes[1].set_title('Social support vs Happiness Score')  # Set title for the plot
axes[1].set_xlabel('Social support')  # Set label for x-axis
axes[1].set_ylabel('Happiness Score')  # Set label for y-axis
# Calculate R-squared value
slope, intercept, r_value, p_value, std_err = linregress(data_numeric['Social support'], data_numeric['Score'])
r_squared = r_value ** 2
axes[1].text(0.1, 0.9, f'R-squared: {r_squared:.2f}', transform=axes[1].transAxes, fontsize=12, bbox=dict(facecolor='lightgray', alpha=0.5))

# Scatter plot with regression line for 'Healthy life expectancy' and 'Happiness Score'
sns.regplot(x='Healthy life expectancy', y='Score', data=data_numeric, ax=axes[2])
axes[2].set_title('Healthy life expectancy vs Happiness Score')  # Set title for the plot
axes[2].set_xlabel('Healthy life expectancy')  # Set label for x-axis
axes[2].set_ylabel('Happiness Score')  # Set label for y-axis
# Calculate R-squared value
slope, intercept, r_value, p_value, std_err = linregress(data_numeric['Healthy life expectancy'], data_numeric['Score'])
r_squared = r_value ** 2
axes[2].text(0.1, 0.9, f'R-squared: {r_squared:.2f}', transform=axes[2].transAxes, fontsize=12, bbox=dict(facecolor='lightgray', alpha=0.5))

plt.savefig('scatterplots.png')  
plt.tight_layout()
plt.show()
```

**Factors of happiness** After plotting the variables that affect happiness the most and their regression lines, GDP per capita, Social support, and Healthy life expectancy, we can deduce that their R-squared is not very high but still significant.
